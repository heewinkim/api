```python
from common.core import PyImage
from src.face_detector import FaceDetector
import matplotlib.pyplot as plt
import cv2
```

```python
detector = FaceDetector()
url='https://media.istockphoto.com/photos/beauty-woman-face-portrait-beautiful-model-girl-with-perfect-fresh-picture-id870365492?k=6&m=870365492&s=612x612&w=0&h=dDKfTkdVsvhavAAwZid_v4BMFAxc5-9w9ByyoQTENsM='
img_cv = PyImage.preprocessing_image(url,'url',1,'cv2')
plt.imshow(img_cv[...,::-1]/255.)
```

```python
result = detector.run(img_cv)

for box in result['faces']:
    x,y,x2,y2 = list(map(int,[box['x'],box['y'],box['xw'],box['yh']]))
    cv2.rectangle(img_cv,(x,y),(x2,y2),(255,0,255),3)

plt.imshow(img_cv[...,::-1]/255.)
plt.show()

from pprint import pprint
pprint(result)
```
